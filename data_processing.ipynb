```python
import pandas as pd
import os
```

```python
def OPEN(filename):
    lines=[]
    with open(filename, "r") as f:
        for line in f:
            line=line.strip()
            lines.append(line)

    return lines


def savefile(filename,LIST):
    with open(filename, 'w') as f:
        for line in LIST:
            f.write("%s\n" % line)
```

```python
def process_headers(s):
    s = s.replace(":"," ")
    new_col_ls = []
    col_list = s.split(";")
    for i,col_name in enumerate(col_list):
        try:
            a = col_name.strip().split(" ")[0]
            new_col_ls.append(a)
        except:
            new_col_ls.append("")
              
    new_header = ";".join(new_col_ls)
    return [new_header]
```

```python
row_limit = 10000
```

# files With header

```python
file_1 = files_with_header[0]
file_1_data_ls = OPEN(os.path.join("data",file_1))
```

```python
len(file_1_data_ls)
```

```python
header = process_headers(file_1_data_ls[0])
new_file_1_data_ls = header+file_1_data_ls[1:]
new_file_1_data_ls = new_file_1_data_ls[:row_limit]
```

```python
new_file_1_data_ls
```

```python
savefile(os.path.join(save_path,file_1),new_file_1_data_ls)
```

```python
file_2= files_without_header[0]
file_2_data_ls_header = OPEN(os.path.join("data",file_2[0]))
file_2_data_ls = OPEN(os.path.join("data",file_2[1]))
```
